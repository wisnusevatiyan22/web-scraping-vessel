```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Fungsi untuk melakukan scraping data dari URL tertentu
def scrape_ship_data(mmsi, imo):
    url = f'https://www.myshiptracking.com/vessels/mmsi-{mmsi}-imo-{imo}'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Mencari tabel dengan kelas tertentu
    table = soup.find('table', class_='table table-sm my-2')

    if table:
        rows = table.find_all('tr')
        data = []
        for row in rows:
            cols1 = row.find_all('th')
            cols1 = [ele.text.strip() for ele in cols1]
            cols2 = row.find_all('td')
            cols2 = [ele.text.strip() for ele in cols2]
            combined_cols = cols1 + cols2  # Menggabungkan kolom header dan data
            data.append(combined_cols)

        # Mengembalikan data sebagai DataFrame sementara
        temp_df = pd.DataFrame(data)
        return temp_df
    return None

# List untuk menyimpan hasil scraping
scraped_data = []

# Loop melalui setiap baris dalam DataFrame
for index, row in df.iterrows():
    imo = int(row['IMO'])
    mmsi = int(row['MMSI'])
    temp_df = scrape_ship_data(mmsi, imo)
    if temp_df is not None:
        scraped_data.append(temp_df)

# Menggabungkan semua DataFrame sementara menjadi satu DataFrame utama
if scraped_data:
    df_scraped = pd.concat(scraped_data, ignore_index=True)
    df_scraped.to_excel('scraped_data.xlsx', index=False)
    print("Scraping selesai dan data disimpan dalam 'scraped_data.xlsx'")
else:
    print("Tidak ada data yang di-scrape.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL website yang ingin di-scrape
base_url = 'https://www.marinevesseltraffic.com/vessels?vessel=&flag=Philippines&page={}&sort=none&direction=none#table-anchor'

# Mengambil konten halaman web
dataframes = []
page = 1
while True:
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Mencari tabel di halaman web
    table = soup.find('table')

    # Jika tabel ditemukan, tambahkan data ke dataframe
    if table:
        rows = []
        for row in table.find_all('tr'):
            rows.append([cell.get_text(strip=True) for cell in row.find_all(['th', 'td'])])

        # Jika tabel memiliki header, gunakan header pertama sebagai nama kolom
        if len(rows) > 0 and all(len(row) == len(rows[0]) for row in rows):
            headers = rows.pop(0)
        else:
            headers = None

        # Membuat dataframe dari nama kolom dan data
        df = pd.DataFrame(rows, columns=headers)
        dataframes.append(df)

        print(f"Scraped page {page}")
        page += 1
    else:
        break

# Menggabungkan semua dataframe menjadi satu
combined_df = pd.concat(dataframes, ignore_index=True)

# Menyimpan dataframe ke dalam satu file Excel
excel_file = "combined_data.xlsx"
combined_df.to_excel(excel_file, index=False)
print(f"Dataframe tersimpan dalam file {excel_file}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Fungsi untuk scraping data dari satu halaman
def scrape_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Sesuaikan sesuai struktur tabel yang ada di website tersebut
    table = soup.find('table')
    rows = table.find_all('tr')

    data = []
    for row in rows[1:]:  # Skip header row
        cols = row.find_all('td')
        cols = [ele.text.strip() for ele in cols]
        data.append(cols)

    return data

# Fungsi untuk mendapatkan halaman terakhir
def get_last_page(base_url):
    response = requests.get(base_url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # Cari elemen yang menunjukkan jumlah total halaman, sesuaikan selector sesuai dengan struktur halaman
    pagination = soup.find('ul', {'class': 'pagination'})
    last_page = 1
    if pagination:
        pages = pagination.find_all('li')
        last_page = int(pages[-2].text)  # Ambil halaman terakhir dari elemen pagination

    return last_page

# Fungsi utama untuk scraping multipages
def scrape_multipages(base_url, start_page, end_page):
    all_data = []
    for page in range(start_page, end_page + 1):
        url = f"{base_url}&page={page}"
        data = scrape_page(url)
        all_data.extend(data)
    return all_data

# URL dasar tanpa parameter halaman
base_url = 'https://www.marinevesseltraffic.com/vessels?flag=ID&type=103&status=Active,%20Commissioned'
start_page = 1
end_page = get_last_page(base_url)  # Mendapatkan halaman terakhir secara otomatis

# Scraping data dari beberapa halaman
scraped_data = scrape_multipages(base_url, start_page, end_page)

# Periksa apakah ada data yang di-scrape
if scraped_data:
    # Membuat DataFrame dan menyimpan ke file Excel
    columns = ['Vessel Name', 'Flag', 'Type', 'Status', 'Column5', 'Column6', 'Column7', 'Column8', 'Column9']  # Ganti dengan nama kolom yang sesuai
    df = pd.DataFrame(scraped_data, columns=columns)
    df.to_excel('scraped_data.xlsx', index=False)
    print("Scraping selesai dan data disimpan di 'scraped_data.xlsx'.")
else:
    print("Tidak ada data yang di-scrape.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL dasar website yang ingin di-scrape
base_url = 'https://www.marinevesseltraffic.com/vessels?page=1&flag=ID&type=103&status=Active,%20Commissioned'

dataframes = []

# Loop melalui halaman-halaman untuk scraping
for page_num in range(1, 6):  # Misalnya, scrap 5 halaman pertama
    url = base_url.format(page_num)

    # Mengambil konten halaman web
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Mencari tabel di halaman web
    tables = soup.find_all('table')

    # Mengambil data dari tabel
    for table in tables:
        rows = []
        for row in table.find_all('tr'):
            rows.append([cell.get_text(strip=True) for cell in row.find_all(['th', 'td'])])

        # Jika tabel memiliki header, gunakan header pertama sebagai nama kolom
        if len(rows) > 0 and all(len(row) == len(rows[0]) for row in rows):
            headers = rows.pop(0)
        else:
            headers = None

        # Membuat dataframe dari nama kolom dan data
        df = pd.DataFrame(rows, columns=headers)
        dataframes.append(df)

# Menyimpan dataframe ke dalam file Excel
for i, df in enumerate(dataframes):
    excel_file = f"dataframe_{i+1}.xlsx"
    df.to_excel(excel_file, index=False)
    print(f"Dataframe {i+1} tersimpan dalam file {excel_file}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL website yang ingin di-scrape
url = 'https://www.marinevesseltraffic.com/vessels?page=1&flag=ID&type=100&status=Active,%20Commissioned'

# Mengambil konten halaman web
response = requests.get(url)
if response.status_code == 200:
    soup = BeautifulSoup(response.content, 'html.parser')

    # Mencari tabel-tabel di halaman web
    tables = soup.find_all('table')

    # Menyimpan data tabel ke dalam list
    dataframes = []
    for table in tables:
        # Mendapatkan data dari setiap baris tabel
        rows = []
        for row in table.find_all('tr'):
            rows.append([cell.get_text(strip=True) for cell in row.find_all(['th', 'td'])])

        # Jika tabel memiliki header, gunakan header pertama sebagai nama kolom
        if len(rows) > 0 and all(len(row) == len(rows[0]) for row in rows):
            headers = rows.pop(0)
        else:
            headers = None

        # Membuat dataframe dari nama kolom dan data
        df = pd.DataFrame(rows, columns=headers)
        dataframes.append(df)

    # Menyimpan dataframe ke dalam file Excel
    with pd.ExcelWriter('marine_vessel_data.xlsx') as writer:
        for i, df in enumerate(dataframes):
            sheet_name = f"Table_{i+1}"
            df.to_excel(writer, sheet_name=sheet_name, index=False)
            print(f"Dataframe {i+1} tersimpan dalam sheet {sheet_name} di file 'marine_vessel_data.xlsx'")
else:
    print(f"Failed to retrieve the web page. Status code: {response.status_code}")
```
