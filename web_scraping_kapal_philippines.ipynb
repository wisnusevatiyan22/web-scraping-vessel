```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL website Baltic Shipping
url = "https://www.balticshipping.com/vessels/"

# Baca daftar nama kapal dari Excel
df_kapal = pd.read_excel("List of shipping companies in philippines 2016.xlsx")

# Buat list kosong untuk menyimpan data
data_kapal = []

# Looping untuk setiap nama kapal
for nama_kapal in df_kapal["NAME OF SHIP"]:
    # Buat request ke URL website
    response = requests.get(url + nama_kapal)

    # Parse HTML dengan BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Temukan tabel spesifikasi kapal
    tabel_spesifikasi = soup.find("table", class_="table ship-info")

    # Jika tabel spesifikasi ditemukan
    if tabel_spesifikasi:
        # Buat dictionary untuk menyimpan spesifikasi
        spesifikasi = {}

        # Looping untuk setiap baris dalam tabel
        for baris in tabel_spesifikasi.find_all("tr"):
            # Temukan kolom dan nilai spesifikasi
            kolom = baris.find("th").text.strip()
            nilai = baris.find("td").text.strip()

            # Simpan kolom dan nilai ke dictionary
            spesifikasi[kolom] = nilai

        # Simpan dictionary spesifikasi ke list
        data_kapal.append(spesifikasi)

# Buat DataFrame Pandas dari list data
df_data_kapal = pd.DataFrame(data_kapal)

# Simpan DataFrame ke file Excel
df_data_kapal.to_excel("data_kapal.xlsx")

# Print hasil
print("Data kapal berhasil disimpan ke file data_kapal.xlsx")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL website Baltic Shipping
url = "https://www.balticshipping.com/vessels/"

# Baca daftar nama kapal dari Excel
df_kapal = pd.read_excel("List of shipping companies in philippines 2016.xlsx")

# Buat list kosong untuk menyimpan data
data_kapal = []

# Looping untuk setiap nama kapal
for nama_kapal in df_kapal["NAME OF SHIP"]:
    # Pisahkan kata pertama dari nama kapal
    nama_kapal_split = nama_kapal.split(" ")

    # Abaikan kata pertama jika bukan huruf
    if not nama_kapal_split[0].isalpha():
        nama_kapal_split = nama_kapal_split[1:]

    # Gabungkan kembali nama kapal
    nama_kapal = " ".join(nama_kapal_split)

    # Buat request ke URL website
    response = requests.get(url + nama_kapal)

    # Parse HTML dengan BeautifulSoup
    soup = BeautifulSoup(response.content, "html.parser")

    # Temukan elemen yang berisi detail kapal
    detail_kapal = soup.find("div", class_="ship-details")

    # Jika detail kapal ditemukan
    if detail_kapal:
        # Buat dictionary untuk menyimpan spesifikasi
        spesifikasi = {}

        # Looping untuk setiap label dan elemen nilai
        for label in detail_kapal.find_all("strong"):
            nilai = label.find_next_sibling().text.strip()
            kolom = label.text.strip().replace(":", "")
            # Simpan kolom dan nilai ke dictionary
            spesifikasi[kolom] = nilai

        # Simpan dictionary spesifikasi ke list
        data_kapal.append(spesifikasi)

# Buat DataFrame Pandas dari list data
df_data_kapal = pd.DataFrame(data_kapal)

# Simpan DataFrame ke file Excel
df_data_kapal.to_excel("data_kapal.xlsx")

# Print hasil
print("Data kapal berhasil disimpan ke file data_kapal.xlsx")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL website Baltic Shipping
url = "https://www.balticshipping.com/vessel/imo/8404264"

# Buat request ke URL website
response = requests.get(url)

# Parse HTML dengan BeautifulSoup
soup = BeautifulSoup(response.content, "html.parser")

# Temukan elemen yang berisi detail kapal
detail_kapal = soup.find("div", class_="ship-details")

# Buat dictionary untuk menyimpan spesifikasi
spesifikasi = {}

# Periksa apakah detail kapal ditemukan
if detail_kapal:
    # Looping untuk setiap label dan elemen nilai
    for label in detail_kapal.find_all("strong"):
        nilai = label.find_next_sibling().text.strip()
        kolom = label.text.strip().replace(":", "")
        # Simpan kolom dan nilai ke dictionary
        spesifikasi[kolom] = nilai
else:
    print("Detail kapal tidak ditemukan. Struktur website mungkin telah berubah.")

# Buat DataFrame Pandas dari dictionary
df_spesifikasi = pd.DataFrame(spesifikasi.items(), columns=["Kolom", "Nilai"])

# Print DataFrame
print(df_spesifikasi.to_string())
```

```python
import requests
from bs4 import BeautifulSoup
import time

# URL website
url = "https://www.balticshipping.com/vessel/imo/6812986"

# Mengirim permintaan ke website
response = requests.get(url)

# Memeriksa status code
if response.status_code == 200:
    html_content = response.content
else:
    print(f"Gagal mengambil konten HTML. Kode status: {response.status_code}")
    exit()

# Parsing HTML dengan BeautifulSoup
soup = BeautifulSoup(html_content, "lxml")

# Menunggu JavaScript dimuat (jika diperlukan)
time.sleep(5)

# Mencari elemen tabel
table_element = soup.find("table", class_="table ship-info")

# Mengekstrak data tabel
if table_element:
    # Mendapatkan header tabel
    headers = [th.text.strip() for th in table_element.find_all("th")]

    # Mendapatkan baris data
    data_rows = []
    for row in table_element.find_all("tr"):
        data_cells = [td.text.strip() for td in row.find_all("td")]
        data_rows.append(data_cells)

    # Menampilkan data tabel
    print("-" * 80)
    print(f"Informasi Kapal - {headers[0]}")
    print("-" * 80)
    for row in data_rows:
        print("| ".join(row))
    print("-" * 80)
else:
    print("Elemen tabel tidak ditemukan.")
```

```python
import requests
from bs4 import BeautifulSoup

# Send HTTP request
url = "https://www.balticshipping.com/vessel/imo/6812986"
try:
    response = requests.get(url)
    response.raise_for_status()  # Raise an exception for non-200 status codes
except requests.exceptions.RequestException as e:
    print("Error fetching website:", e)
    exit()

# Parse HTML content
try:
    soup = BeautifulSoup(response.content, "html.parser")
    table = soup.find("table", class_="table ship-info")
except AttributeError:
    print("Table element not found on the webpage.")
    exit()

# Extract data (assuming table was found)
if table:
    data = {}
    for row in table.find_all("tr", recursive=False):
        th = row.find("th")
        td = row.find("td")
        if th and td:
            key = th.text.strip()
            value = td.text.strip()
            if key:
                data[key] = value
    print(data)
else:
    print("Couldn't extract data from the table.")
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.balticshipping.com/vessel/imo/6812986'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')

ship_info_table = soup.find('table', class_='table ship-info')
ship_info = {}
for row in ship_info_table.find_all('tr'):
    columns = row.find_all(['th', 'td'])
    if len(columns) == 2:
        key = columns[0].text.strip()
        value = columns[1].text.strip()
        ship_info[key] = value

# Print the extracted information
for key, value in ship_info.items():
    print(f"{key}{value}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Inisialisasi list untuk menyimpan informasi kapal
ships = []

# Iterasi melalui halaman-halaman web yang akan di-scrape (dalam contoh ini, 5 halaman)
for i in range(1, 6):
    url = f"https://www.balticshipping.com/vessels"  # Ganti dengan URL halaman web yang sesuai
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Mendapatkan semua elemen <a> dengan kelas 'vessel-list-item'
    vessels = soup.find_all('a', class_='col-sm-4 vessel-list-item')

    # Iterasi melalui setiap elemen <a> untuk mendapatkan informasi kapal
    for vessel in vessels:
        name = vessel.find('h4').text.strip()
        imo_number = vessel['href'].split('/')[-1]
        vessel_type = vessel.find('h5').text.strip()
        image_url = vessel.find('img')['src']
        flag_code = vessel.find('i')['class'][1].split('-')[-1]

        # Menambahkan informasi kapal ke dalam list
        ships.append({
            'Name': name,
            'IMO Number': imo_number,
            'Type': vessel_type,
            'Image URL': image_url,
            'Flag Code': flag_code
        })

# Membuat DataFrame dari list ships
df = pd.DataFrame(ships, columns=['Name', 'IMO Number', 'Type', 'Image URL', 'Flag Code'])

# Menyimpan DataFrame ke file CSV
df.to_csv('ships.csv', index=False)
```
