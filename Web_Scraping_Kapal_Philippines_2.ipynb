```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL website yang ingin di-scrape
base_url = 'https://www.marinevesseltraffic.com/vessels?vessel=&flag=Philippines&page={}&sort=none&direction=none#table-anchor'

# Mengambil konten halaman web
dataframes = []
page = 1
while True:
    url = base_url.format(page)
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Mencari tabel di halaman web
    table = soup.find('table')

    # Jika tabel ditemukan, tambahkan data ke dataframe
    if table:
        rows = []
        for row in table.find_all('tr'):
            rows.append([cell.get_text(strip=True) for cell in row.find_all(['th', 'td'])])

        # Jika tabel memiliki header, gunakan header pertama sebagai nama kolom
        if len(rows) > 0 and all(len(row) == len(rows[0]) for row in rows):
            headers = rows.pop(0)
        else:
            headers = None

        # Membuat dataframe dari nama kolom dan data
        df = pd.DataFrame(rows, columns=headers)
        dataframes.append(df)

        print(f"Scraped page {page}")
        page += 1
    else:
        break

# Menggabungkan semua dataframe menjadi satu
combined_df = pd.concat(dataframes, ignore_index=True)

# Menyimpan dataframe ke dalam satu file Excel
excel_file = "combined_data.xlsx"
combined_df.to_excel(excel_file, index=False)
print(f"Dataframe tersimpan dalam file {excel_file}")
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Load data list MMSI dan IMO dari file Excel
data = pd.read_excel('combined_data.xlsx')

# Fungsi untuk melakukan scraping dan menyimpan data ke dalam dataframe
def scrape_data(mmsi, imo):
    url = f"https://www.myshiptracking.com/vessels/mmsi-{mmsi}-imo-{imo}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    table = soup.find('table', class_='myst-table table m-0')
    if table:
        rows = []
        for row in table.find_all('tr'):
            cells = row.find_all('td')
            if len(cells) == 2:
                rows.append({
                    'Port': cells[0].get_text(strip=True),
                    'Arrivals': cells[1].get_text(strip=True)
                })
        df = pd.DataFrame(rows)
        return df
    return None

# Looping untuk scraping dan menggabungkan data dari setiap MMSI dan IMO
dataframes = []
for index, row in data.iterrows():
    mmsi = row['MMSI⇅']
    imo = row['IMO⇅']
    df = scrape_data(mmsi, imo)
    if df is not None:
        dataframes.append(df)
        print(f"Scraped data for MMSI: {mmsi}, IMO: {imo}")

# Menggabungkan semua dataframe menjadi satu
combined_df = pd.concat(dataframes, ignore_index=True)

# Menyimpan dataframe ke dalam satu file Excel
excel_file = "port arrivals.xlsx"
combined_df.to_excel(excel_file, index=False)
print(f"Dataframe tersimpan dalam file {excel_file}")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Daftar MMSI dan IMO untuk setiap kapal
data = pd.read_excel('combined_data.xlsx')

# URL template untuk web scraping
url_template = 'https://www.myshiptracking.com/vessels/mmsi-{mmsi}-imo-{imo}'

# Inisialisasi DataFrame kosong untuk menyimpan hasil scraping
results = []

for kapal in data:
    # Mengganti kode MMSI dan IMO dalam URL template
    url = url_template.format(mmsi=kapal['MMSI⇅'], imo=kapal['IMO⇅'])

    # Mengambil konten halaman web
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Mencari tabel dengan kelas 'myst-table table m-0'
    table = soup.find('table', class_='myst-table table m-0')

    # Jika tabel ditemukan
    if table:
        # Membaca tabel ke dalam DataFrame pandas
        df = pd.read_html(str(table))[0]

        # Menambahkan kolom 'nama_kapal' ke DataFrame
        df['nama_kapal'] = kapal['Vessel name⇅']

        # Menambahkan DataFrame ke dalam results
        results.append(df)
    else:
        print(f"Tabel tidak ditemukan untuk kapal {kapal['Vessel name⇅']}")

# Menggabungkan semua hasil scraping menjadi satu DataFrame
result_df = pd.concat(results, ignore_index=True)

# Menyimpan hasil scraping ke dalam file Excel
result_df.to_excel('hasil_scraping_kapal.xlsx', index=False)

# Menampilkan hasil scraping
print(result_df)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Nomor IMO kapal yang akan dicari
imo_number = '7911193'

# URL website pencarian
url = f'https://www.igpandi.org/vessel-search/{imo_number}'

# Mengirim permintaan GET ke URL
response = requests.get(url)

# Mengecek apakah permintaan berhasil
if response.status_code == 200:
    # Menginisialisasi objek BeautifulSoup
    soup = BeautifulSoup(response.text, 'html.parser')

    # Mencari div dengan class 'vessel-search-result'
    result_div = soup.find('div', class_='result-box')

    # Mengekstrak informasi dari div
    name = result_div.find('h3').text.strip()
    details = result_div.find_all('p')
    details_text = [detail.text.strip() for detail in details]

    # Menyimpan informasi ke dalam DataFrame
    df = pd.DataFrame({'Detail': details_text})

    # Menambahkan nama kapal sebagai kolom tambahan
    df.insert(0, 'Name', name)

    # Menyimpan DataFrame ke file Excel
    excel_file = f'vessel_search_{imo_number}.xlsx'
    df.to_excel(excel_file, index=False)

    print(f'Hasil pencarian untuk nomor IMO {imo_number} disimpan dalam file {excel_file}')
else:
    print('Gagal melakukan web scraping')
```
